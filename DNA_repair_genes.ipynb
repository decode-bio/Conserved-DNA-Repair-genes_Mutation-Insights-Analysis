```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Now Coverting the TSV(Tab separated values) files into the CSV(Comma separated values) for further analysis

# PMS2 gene Mutaion analysis

```python
pd.read_csv('Mutation_data_pms2.tsv', sep='\t').to_csv('Mutation_data_pms2.csv', index=False)
mut_data_pms2=pd.read_csv('Mutation_data_pms2.csv')
```

```python
mut_data_pms2
```

# As we will not be using all the values in the given table we can ignore the null values as it will not affect out analysis

```python
Cancer = mut_data_pms2['Cancer Type']
Frequency = mut_data_pms2['Allele Freq (T)']
```

```python
plt.figure(figsize=(10,6))
plt.bar(Cancer, Frequency, color='skyblue', edgecolor='black')
plt.title('Cancer type vs. Allele frequency')
plt.xlabel('Cancer Type')
plt.ylabel('Allele Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
Cancer = mut_data_pms2['Cancer Type']
Mutaion = mut_data_pms2['# Mut in Sample']
```

```python
plt.figure(figsize=(10,6))
plt.bar(Cancer, Mutaion, color='orange', edgecolor='black')
plt.title('Cancer type vs. Mutaions in PMS2 ')
plt.xlabel('Cancer Type')
plt.ylabel('Mutations')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Rad51c gene mutation Analysis

```python
pd.read_csv('Mutation_data_rad51c.tsv', sep='\t').to_csv('Mutation_data_rad51c.csv', index=False)
mut_data_rad51c=pd.read_csv('Mutation_data_rad51c.csv')
```

```python
mut_data_rad51c
```

```python
Cancer2 = mut_data_rad51c['Cancer Type']
Frequency2 = mut_data_rad51c['Allele Freq (T)']
```

```python
plt.figure(figsize=(14,6))
plt.bar(Cancer2, Frequency2, color='skyblue', edgecolor='black')
plt.title('Cancer type vs. Allele frequency for rad51c gene')
plt.xlabel('Cancer Type')
plt.ylabel('Allele Frequency')
plt.xticks(rotation=45, ha='right') 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
Cancer1 = mut_data_rad51c['Cancer Type']
Mutaion1 = mut_data_rad51c['# Mut in Sample']
```

```python
plt.figure(figsize=(14,6))
plt.bar(Cancer1, Mutaion1, color='orange', edgecolor='black')
plt.title('Cancer type vs. Mutation for rad51c gene')
plt.xlabel('Cancer Type')
plt.ylabel('Mutation')
plt.xticks(rotation=45, ha='right') 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# lets to see in which cancer type the gene has mutated the most 
type_counts = mut_data_rad51c.groupby('Cancer Type').size()
```

```python
plt.figure(figsize=(10,6))
type_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Cancer Type Occurrence per Sample for rad51c gene')
plt.xlabel(' Cancer Type')
plt.ylabel('Count')
plt.yticks(np.arange(0, 30, 2))
plt.xticks(rotation=45, ha = 'right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
